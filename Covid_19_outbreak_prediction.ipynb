```python
from fbprophet import Prophet
```

```python
df=pd.read_csv("covid_19_india.csv",parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
deaths    = df.groupby('Date').sum()['Deaths'].reset_index()
recovered = df.groupby('Date').sum()['Cured'].reset_index()
```

```python
confirmed = confirmed.rename(columns={'Date': 'ds', 'Confirmed': 'y'})
confirmed.head()
```

```python
m=Prophet(interval_width=0.95)
m.fit(confirmed)
future= m.make_future_dataframe(periods=7)
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
confirmed_plot=m.plot(forecast)
```

```python
confirmed_forecast_plot=m.plot_components(forecast)
```

```python
deaths = deaths.rename(columns={'Date': 'ds', 'Deaths': 'y'})
deaths.tail()
```

```python
m=Prophet(interval_width=0.95)
m.fit(deaths)
future=m.make_future_dataframe(periods=7)
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
confirmed_deaths_plot=m.plot(forecast)
```

```python
deaths_forecast_plot=m.plot_components(forecast)
```

```python
recovered = recovered.rename(columns={'Date': 'ds', 'Cured': 'y'})
deaths.tail()
```

```python
m=Prophet(interval_width=0.95)
m.fit(recovered)
future=m.make_future_dataframe(periods=7)
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
confirmed_recovered_plot=m.plot(forecast)
```

```python
recovered_forecast_plot=m.plot_components(forecast)
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(m, initial='50 day', period='180 days', horizon = '365 days')
```

```python
df_cv
```

```python
from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```
